# Suicides in India (2001 - 2012)

In this notebook, we analyse the dataset of suicides in India from 2001 to 2012. We will analyse the data using a datset from Kaggle.

First we will import all the libraries required to analyse the dataset.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data/data.csv')
data.head()
```

```python
data = data.drop(data[data.Total == 0].index)
```

```python
data.State.value_counts()
```

```python
data.Year.value_counts()
```

```python
data.Type_code.value_counts()
```

```python
data.Gender.value_counts()
```

```python
data.Age_group.value_counts()
```

```python
data = data.drop(data[data.Age_group == '0-100+'].index)
```

```python
gender = data.Gender.value_counts()
gender
```

```python
gender.plot(kind = 'bar', x = 'Gender', y = 'Total', xlabel = 'Gender', ylabel = 'Suicides', title = 'Suicides by Gender')
```

```python
states = data.State.value_counts()[data.State.unique()]
states
```

```python
states.plot(kind = 'bar', x = 'State', y = 'Total', xlabel = 'States', ylabel = 'Suicides', title = 'Suicides in States')
```

```python
type_codes = data.Type_code.value_counts()[data.Type_code.unique()]
type_codes
```

```python
type_codes.plot(kind = 'bar', x = 'Type_code', y = 'Total', xlabel = 'Type', ylabel = 'Suicides', title = 'Suicides by Type')
```

```python
ages = data.Age_group.value_counts()[data.Age_group.unique()]
ages
```

```python
ages.plot(kind= 'bar', x = 'Age_group', y = 'Total', xlabel = 'Age-Groups', ylabel = 'Suicides', title = 'Suicies by Age-Groups')
```

```python
plt.plot(ages)
plt.xlabel('Age-Groups')
plt.ylabel('Suicides')
plt.title('Suicides by Age-Groups')
plt.show()
```

```python
years = data.Year.value_counts()[data.Year.unique()]
years
```

```python
plt.plot(years)
plt.xlabel('Years')
plt.ylabel('Suicides')
plt.title('Suicides by Years')
plt.show()
```
